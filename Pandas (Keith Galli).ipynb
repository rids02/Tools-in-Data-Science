```python
#To geolocate a query to an address and coordinates:
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='user_agent')
location = geolocator.geocode("London")

print(location.address)
print('\n', (location.latitude, location.longitude))
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='user_agent')
location = geolocator.geocode("Chitradurga")
print(location.address)
print('\n', (location.latitude, location.longitude))
```

```python
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='user_agent')

a = geolocator.geocode("Junagadh")
b = geolocator.geocode("Sagar")
am = (a.latitude, a.longitude)
destb = (b.latitude, b.longitude)
print(geodesic(am, destb).km)

c = geolocator.geocode("Karimganj")
d = geolocator.geocode("Pathanamthitta")
destc = (c.latitude, c.longitude)
destd = (d.latitude, d.longitude)
print(geodesic(destc, destd).km)

e = geolocator.geocode("Nagapattinam")
f = geolocator.geocode("Kalaburagi")
deste = (e.latitude, e.longitude)
destf = (f.latitude, f.longitude)
print(geodesic(deste, destf).km)

a = geolocator.geocode("Rewa")
b = geolocator.geocode("Datia")
am = (a.latitude, a.longitude)
destb = (b.latitude, b.longitude)
print(geodesic(am, destb).km)
```

```python
#To find the address corresponding to a set of coordinates:
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='user_agent')
location = geolocator.reverse("52.509669, 13.376294")
print(location.address)
print('\n', (location.latitude, location.longitude))
```

```python
from functools import partial
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent='user_agent')

geocode = partial(geolocator.geocode, language='es')
print(geocode('London'))

geocode = partial(geolocator.geocode, language='en')
print(geocode('London'))

reverse = partial(geolocator.reverse, language='es')
print(reverse("52.509669, 13.376294"))
```
